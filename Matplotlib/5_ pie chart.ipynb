```python
import matplotlib.pyplot as plt
```

#**Drawing pie chart for home expenses**
* **we can use numpy array and panda dataframe to plt the the plots rather than passing list names**

```python
expense_vals = [1400,600,300,410,250]
expense_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(expense_vals,labels=expense_labels)
plt.show()
```

#**Draw pie chart as perfect circle using plt.axis("equal") **

```python
plt.pie(expense_vals,labels=expense_labels, shadow=True)
plt.axis("equal")
plt.show()
```

#**Show percentages for every pie by using autopct='%.2f%%' to show two decimal place and Specifying radius to increase chart size**
* **by using shadow=True we will see shadow of the graph**

```python
plt.axis("equal")
plt.pie(expense_vals,labels=expense_labels, shadow=True, autopct='%.2f%%',radius=1.5)
plt.show()
```

#**Using explode to take a specific pie**

```python
plt.axis("equal")
plt.pie(expense_vals,labels=expense_labels, shadow=True, autopct='%.2f%%',radius=1.5,explode=[0,0.5,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(expense_vals,labels=expense_labels, shadow=True, autopct='%.2f%%',radius=1.5,explode=[0,0,0.2,0,0])
plt.show()
```

#**counterclock and angle properties**

```python
plt.axis("equal")
plt.pie(expense_vals,labels=expense_labels, shadow=True, autopct='%.2f%%',radius=1.5,explode=[0,0,0.2,0,0], startangle=180)
plt.show()
```

```python
plt.axis("equal")
plt.pie(expense_vals,labels=expense_labels, shadow=True, autopct='%.2f%%',radius=1.5,explode=[0,0,0.2,0,0],counterclock=True, startangle=45)
plt.show()
```

#**We can use plt.savefig('location/graph name.format') to save the graph**

```python
plt.savefig('piechart.png' ,bbox_inches='tight',pad_inches=2,transparent=True)
```
