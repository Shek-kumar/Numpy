#**using python dictionary to create a dataframe**

```python
import pandas as pd 
wheather_data={
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']
}
df = pd.DataFrame(wheather_data)

df
```

#**getting the size of the dataframe**

```python
print(df.shape)
rows,columns=df.shape
print("no of rows is:",rows)
print("no of cols:",columns)
```

#**using head(no.of rows you want to print(N)) to print first N rows of the dataframe**

```python
print(df.head(3))
print("\nWithout passing any argument in the head function \n")
print(df.head())
```

#**using tail(no.of rows you want to print(N)) to print last N rows of the dataframe**

```python
print(df.tail())
print("\n                                 Printing last 3 rows:")
df.tail(3)
```
