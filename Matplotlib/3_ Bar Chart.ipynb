```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['Google','Amazon','FB','Microsoft']
revenue=[90,136,36,250]
profit=[10,20,6,34]
```

#**Since plt do not plots the strings so we assigns each company as numerical value by creating a xpos(xposition) numpy array where first no denotes first company and so on**



```python
xpos=np.arange(len(company))
xpos
```

**plotting the graph for the revenue of company where company is shown by a numerical value**

```python
plt.bar(xpos,revenue)
```

#**to replace the numerical values by their corresponding company's name we use xticks() methods**

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue in billion')
plt.title('Us comapnies revenue for a year')
plt.bar(xpos,revenue,label='revenue')

plt.legend()
```

#**Plotting two bar chart in a single graph**

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue in billion')
plt.title('Us comapnies revenue for a year')
plt.bar(xpos,revenue,label='revenue')
plt.bar(xpos,profit,label='profit')
plt.legend()
```

#**Above we obtained both the bar chart at same place to obtain them at different places we proceed as following**

* **we can also change the width of the bar chart**

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue in billion')
plt.title('Us comapnies revenue for a year')
plt.bar(xpos-0.2,revenue,label='revenue',width=0.4)
plt.bar(xpos+0.2,profit,label='profit',width=0.4)
plt.legend()
```

#**To get the horizontal barchart we use the method barh()**

```python
plt.yticks(xpos,company)
plt.xlabel('Revenue in billion')
plt.title('Us comapnies revenue for a year')

plt.barh(xpos-0.2,revenue,label='revenue')
plt.barh(xpos+0.2,profit,label='profit')

plt.legend()
```
