```python
df=pd.read_csv('weather_by_cities.csv')
df
```

#using pivot in order to get city as index and date as columns

```python
df.pivot(index='city',columns='day')
```

#using pivot in order to get day as index and city  as columns and only temperature as the values

```python
df.pivot(index='day',columns='city',values='temperature')
```

#**Pivot table is uded to summarize and aggregate the data inside dataframe**

```python
df=pd.DataFrame({
    'day':'01/05/2017',
    'city':['mumbai','mumbai','delhi','delhi'],
    'temp':[32,25,32,40],
    'humidity':[10,32,65,82]
})  
df
```

#using pivot_index in order to get avg temp for a city on a specific day

```python

df.pivot_table(index="city",columns="day")
```

#default valuye of aggfunc is mean 


1.   we can use diif sum count and other function in aggfunc



```python
df.pivot_table(index="city",columns="day",aggfunc='sum')
```

#**Margins**  It add all row / columns (e.g. for subtotal / grand totals).

```python
import numpy as np
df.pivot_table(index="city",columns="day", margins=True,aggfunc='sum')
```

#**Grouper**

```python
df1111=pd.DataFrame({
    'date':['01/05/2017','02/05/2017','03/05/2017','01/06/2017','02/06/2017','03/06/2017'],
    'city':['mumbai','mumbai','mumbai','mumbai','mumbai','mumbai'],
    'temp':[32,25,32,40,23,50],
    'humidity':[10,32,65,82,65,97]
})  
df1111
```

#converting date in date format which is actuallly in string format

```python
df1111['date'] = pd.to_datetime(df1111['date'])
```

#calculating monthly avg temp

```python

df1111.pivot_table(index=pd.Grouper(freq='M',key='date'),columns='city')
```
