#creating two dataframes using dictionary to concante them

```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

#Basic Concatenation of two dataframes **Using concat() functions**

```python
df = pd.concat([india_weather, us_weather])
df
```

#Ignoring Index **by default ignore_index=False**

```python
df = pd.concat([india_weather, us_weather],ignore_index=True)
df
```

  #**Concatenation And Keys**

```python
df = pd.concat([india_weather, us_weather],keys=['India','Us'])
df
```

#**getting the dataframe Of India dats using the index India we created above**

```python
df.loc["India"]
```

#Adding the data of the second dataframe as column in the first dataframe 

```python
# creating dataframes
temp_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30]
},index=[0,1,2])
temp_df
```

```python
# creating dataframe 
wind_df = pd.DataFrame({
    "city": ["delhi","mumbai","banglore"],
    "windspeed": [12,5,10]
},index=[1,0,2])
wind_df
```

```python
df=pd.concat([temp_df,wind_df])
df
```

#here we appended the two dataframes as above then we didn't get the desired output so we will use axis =1 to add the data in desired format

```python
df=pd.concat([temp_df,wind_df],axis=1)
df
```
