```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,12,35,23,22,38,32]
```

#**Plotting First graph using matplotlib.pylot method**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,color='green',linewidth=5,linestyle='dashed')
```

#**Plotting graph such that color of point is green and denoted by a + or * sign**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,'g*')
```

#**Using -- for the dotted graph**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,'g+--')
```

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,'--*g')                        #here order doesn't matters we can write '--*g' or'--g*' or 'g*--' or 'g--*' anything
```

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,'gD')                          #order doesn't matters
```

#**Getting same type of graph by assigning value to color,assigning markertype,linestyle=''**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,color='green',marker='D',linestyle='')
```

#**Using markersize=N to control the size of the marker**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,color='green',marker='D',linestyle='',markersize=10)
```

#**alpha property of plot API**

**alpha function is used for transparency of the plot**


1.   **alpha=1 means completely opaque**
2.   **alpha=0 means completely transparent**



```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')
plt.plot(x,y,color='red',alpha=1)
```
