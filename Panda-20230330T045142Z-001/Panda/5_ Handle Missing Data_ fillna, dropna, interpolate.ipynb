#**making the type of the day= date rather than string**

```python
df=pd.read_csv('weather_data_NEW_YORK.csv',parse_dates=['day'])
print("Type of the day now is:",type(df.day[0]))
df
```

#**setting day as index**

```python
df.set_index('day',inplace=True)
df
```

#**using fillna() method to fill the NaN values with the value of our choice**

```python
new_df=df.fillna(0)
new_df
```

#**filling the missing values with the value of our choice columnwise by using dictionary**

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':1,
    'event':"no event"
})
new_df
```

#**using method='ffill' (forward fill) to fill the Nan values with the previous day value**

```python
new_df=df.fillna(method='ffill')
new_df
```

#**using method='bfill' (forward fill) to fill the Nan values with the next day value**

```python
new_df=df.fillna(method='bfill')
new_df
```

#**use of axis in fillna**

#1.   axis=1 denotes column it will copy the values from next cell



```python
new_df=df.fillna(method='bfill',axis=1)
new_df
```

#axis=0 means row it will works as originally the file was working

```python
new_df=df.fillna(method='bfill',axis=0)
new_df
```

#**Use of limit in fillna()**

#To copy the value from the previous/next cell upto a fixed no.(say n) of row we use limit=n 

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```

```python
new_df=df.fillna(method='bfill',limit=2)
new_df
```

#**Using interpolate() method to fill missing values with some approximate values**

1.  By default the method is linear




```python
new_df=df.interpolate()
new_df
```

#**using method=time in interpolate method to make more accurate approximation accordingingly to day**

```python
new_df=df.interpolate(method='time')
new_df
```

#**Dropna method**

#dropping all the rows in which any of the three value is missing

```python
new_df=df.dropna()
new_df
```

#dropping all the rows in which all of the three value is missing


```python
new_df=df.dropna(how='all')
new_df
```

#using thresh=n to drop those rows which has atleast n non missing values 

```python
new_df=df.dropna(thresh=1)
new_df
```

```python
new_df=df.dropna(thresh=2)
new_df
```
