#Crosstab (also known as contingency table or cross tabulation) is a table showing frequency distribution of one variable in rows and another on columns. pandas crosstab method can be used to generate these contingency tables that are extremely useful in survey and business analytics.

#**using crosstab to build the frequency table**

```python

pd.crosstab(df.Nationality,df.Handedness)
```

```python

pd.crosstab(df.Sex,df.Handedness)
```

#using margin to get the total

```python

pd.crosstab(df.Nationality,df.Handedness,margins='True')
```

#**Multi Index Column and Rows**

```python
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins=True)
```

```python
pd.crosstab([df.Nationality,df.Sex], df.Handedness, margins=True)
```

#**Normalize** to get the percentage 

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index')
```
