#**Resampling**

#calculating average monthly closing price of apple stocks

```python
df['Close'].resample('M').mean().head()
```

#Graph for the avg monthly closing price

```python
%matplotlib inline
df['Close'].resample('M').mean().head().plot()
```

#Graph for quarterly avg stocks price

```python
%matplotlib inline
df['Close'].resample('Q').mean().head().plot()
```

#Bar chart for weekly avg price 

```python
%matplotlib inline
df['Close'].resample('W').mean().head().plot(kind='bar')
```
