#**using a external flag to print the elements of the array column-wise**

```python
for x in np.nditer(a,order='F',flags=['external_loop']):
  print(x)
```

#**using op_flags to Modify array values while iterating**

```python
import numpy as np
a=np.arange(12).reshape(3,4)
print(a)

for x in np.nditer(a,op_flags=['readwrite']):
  x[...]=x**2

print(a)
```
