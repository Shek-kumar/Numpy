arange() is similar to the range

```python
# LESS Memory  requirement of numpy
import numpy as np
import time
import sys

L=range(1000)
print(sys.getsizeof(5)*len(L))

array=np.arange(1000)
print(array.size*array.itemsize)
```

here in above examople the size of one list element is 28 bytes while that of the numpy array is 8 bytes this is because in list firstly it stores the pointer for every values then store the values in the corresponding address

```python
# numpy array is fast and convinient

import numpy as np
import time
import sys

size=1111111
l1=range(size)
l2=range(size)

a1=np.arange(size)
a2=np.arange(size)

start_time=time.time()
result=[(x+y) for x,y in zip(l1,l2)]
print("time taken by python list to compute is: ",(time.time()-start_time)*1000)

start=time.time()
result=a1+a2
print("numpy array took :",(time.time()-start)*1000)
```
