```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
max_temp=[48,38,50,44,40,36,32]
min_temp=[22,20,19,18,23,25,26]
avg_temp=[30,32,36,28,35,36,38]
```

#**Using legend() and label() methods to determine that which graph is for which set in case of when we have more than 1 plots in a single graph**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')

plt.plot(days,max_temp,label="Maximum Temperature")
plt.plot(days,min_temp,label="Minimum Temperature")
plt.plot(days,avg_temp,label="Average Temperature")

plt.legend()
```

#**To move the legend to the desired location we use plt.legend(loc='place where we want to shift it')**


* **Valid font size are xx-small, x-small, small, medium, large, x-large, xx-large, larger, smaller, None**
* **using Shodow=True to make the legend 3D look**  
* **by default the legend is at the best place**



```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')

plt.plot(days,max_temp,label="Maximum Temperature")
plt.plot(days,min_temp,label="Minimum Temperature")
plt.plot(days,avg_temp,label="Average Temperature")

plt.legend(loc='best',shadow=True,fontsize='medium')
```

#**Using grid() method to add the grip to our graph**

```python
plt.xlabel('Day',fontsize=20)
plt.ylabel('Temperature',fontsize=20)
plt.title('Weather data',fontsize='20')

plt.plot(days,max_temp,label="Maximum Temperature")
plt.plot(days,min_temp,label="Minimum Temperature")
plt.plot(days,avg_temp,label="Average Temperature")

plt.legend(loc='best',shadow=True,fontsize='medium')
plt.grid()
```
