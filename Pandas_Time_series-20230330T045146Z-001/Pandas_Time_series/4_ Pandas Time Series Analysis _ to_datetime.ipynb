#handling different types of string format which represent a same time

```python
import pandas as pd
dates = ['2017-01-05', 'Jan 5, 2017', '01/05/2017', '2017.01.05', '2017/01/05','20170105']
pd.to_datetime(dates)
```

#to_datetime() also handles the time

```python
dt = ['2017-01-05 2:30:00 PM', 'Jan 5, 2017 14:30:00', '01/05/2016', '2017.01.05', '2017/01/05','20170105']
pd.to_datetime(dt)
```

#**European style dates with day first**
#yyyy-mm-dd

```python

pd.to_datetime('30-12-2016')
```

#yyyy-mm-dd

```python
pd.to_datetime('5-1-2016', dayfirst=True)
```

#**Custom date time format**

```python
pd.to_datetime('2017$01$05', format='%Y$%m$%d')
```

```python
pd.to_datetime('2017#01#05', format='%Y#%m#%d')
```

#**Handling invalid dates**

```python
pd.to_datetime(['2017-01-05', 'Jan 6, 2017', 'abc'], errors='ignore')   #here no conversion takes place when we set ignore in the errors
```

#If we write errors='coerce' then it will convert all correct date into datetime format leaving the incorrect one

```python
pd.to_datetime(['2017-01-05', 'Jan 6, 2017', 'abc'], errors='coerce')
```

#**Epoch::**Epoch or Unix time means number of seconds that have passed since Jan 1, 1970 00:00:00 UTC time




#converting time which is the total no of seconds passed from jan 1 1970 into the datetime format

```python
time=1834567891
dt=pd.to_datetime([time],unit='s')   # or   dt=pd.to_datetime(time,unit='s')

dt
```
