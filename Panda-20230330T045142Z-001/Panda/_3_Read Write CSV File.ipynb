# **Clean up messy data from file "not available" and "n.a." replace with "na_values"**

```python
df = pd.read_csv("stock_data.csv", na_values=["n.a.", "not available"])
df
```

#**Supply dictionary for replace with "na_values"**

```python
df = pd.read_csv("stock_data.csv",  na_values={
        'eps': ['not available','n.a.'],
        'revenue':['not available','n.a.',-1],
        'people': ['not available','n.a.']           # it means that in price when we encounter n.a or not available then replace that with na_values
    })
df
```



#**Write to CSV**




```python
df.to_csv('new.csv')    # this will create a csv file named as new
```
